## 1. Python Data Types

```python
# Integer, Float, String, Boolean
a = 10          # int
b = 3.14        # float
c = 'Hello'     # string
d = True        # boolean

print(type(a), type(b), type(c), type(d))
```

## 2. Variables

```python
# Variable assignment
x = 5
y = x + 10
print('x:', x, 'y:', y)
```

## 3. Control Flow

```python
# If-elif-else example
num = 7
if num > 10:
    print('Greater than 10')
elif num == 10:
    print('Equal to 10')
else:
    print('Less than 10')
```

```python
# For loop example
for i in range(1, 6):
    print(i)
```

```python
# While loop example
count = 0
while count < 3:
    print('Count:', count)
    count += 1
```

## 4. Collections

```python
# List
fruits = ['apple', 'banana', 'cherry']
fruits.append('orange')
print(fruits)
```

```python
# Tuple (immutable)
point = (10, 20)
print(point[0], point[1])
```

```python
# Set (unique values)
nums = {1, 2, 2, 3}
nums.add(4)
print(nums)
```

```python
# Dictionary (key-value pairs)
student = {'name': 'Alex', 'age': 21}
student['course'] = 'Python'
print(student)
```

## 5. Common Collection Methods

```python
# List methods
numbers = [1, 2, 3]
numbers.extend([4, 5])
numbers.remove(2)
print(numbers)
```

```python
# Dictionary methods
print(student.keys())
print(student.values())
print(student.items())
```

## 6. Functions

```python
# Function definition
def add(a, b):
    """Returns sum of two numbers"""
    return a + b

print(add(3, 4))
```

## 7. Error Handling

```python
# Try-except example
try:
    result = 10 / 0
except ZeroDivisionError:
    print('Cannot divide by zero')
finally:
    print('Execution complete')
```

## 8. Common Python Modules

```python
# math module
import math
print(math.sqrt(16))
```

```python
# datetime module
from datetime import datetime
now = datetime.now()
print(now)
```

```python
# random module
import random
print(random.randint(1, 10))
```
