# Frequently Asked Questions with Solutions

## Reverse a string

```
s = 'python'
print(s[::-1])
```

## Find duplicates in list

```
lst = [1,2,3,2,4,1]
dups = set([x for x in lst if lst.count(x) > 1])
print(dups)
```

## Fibonacci using DP

```
def fib(n):
    dp = [0,1]
    for i in range(2, n+1):
        dp.append(dp[i-1] + dp[i-2])
    return dp[n]

print(fib(10))
```
