# Notebook 5: OOP, Decorators, Generators

```
# OOP example
class Person:
    def __init__(self, name):
        self.name = name
    def greet(self):
        return f'Hello {self.name}'

p = Person('Alex')
print(p.greet())
```

```
# Decorator
def my_decorator(func):
    def wrapper():
        print('Before')
        func()
        print('After')
    return wrapper

@my_decorator
def say_hi():
    print('Hi')

say_hi()
```

```
# Generator
def count_up(n):
    for i in range(n):
        yield i

for x in count_up(3):
    print(x)
```
