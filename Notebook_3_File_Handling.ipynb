# Notebook 3: File Handling, JSON, CSV, Pickle

```
# Text file write & read
with open('sample.txt', 'w') as f:
    f.write('Hello File')

with open('sample.txt', 'r') as f:
    print(f.read())
```

```
# JSON handling
import json
data = {'name': 'Alex', 'age': 30}
with open('data.json', 'w') as f:
    json.dump(data, f)

with open('data.json') as f:
    print(json.load(f))
```

```
# CSV handling
import csv
with open('data.csv', 'w', newline='') as f:
    writer = csv.writer(f)
    writer.writerow(['name', 'score'])
    writer.writerow(['Sam', 90])
```

```
# Pickle (object serialization)
import pickle
obj = {'x': 1}
with open('obj.pkl', 'wb') as f:
    pickle.dump(obj, f)

with open('obj.pkl', 'rb') as f:
    print(pickle.load(f))
```
